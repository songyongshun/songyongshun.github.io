1. 使用Python的sqlite3模块创建一个SQLite数据库，并创建一个名为"employees"的表，包含以下字段：id（整数，主键），name（文本），age（整数），department（文本）。插入三条员工记录，并查询所有记录，打印输出。

注意，如果是第二次运行，记得对数据库文件改名，数据库不允许写入id相同的数据。

```python
import sqlite3
# 连接到SQLite数据库（如果数据库不存在则会创建一个新的数据库文件）
conn = sqlite3.connect('company.db')  
cursor = conn.cursor()
# 创建employees表
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    department TEXT
)
''')
# 插入三条员工记录
employees = [
    (1, 'Alice', 30, 'HR'),
    (2, 'Bob', 25, 'Engineering'),
    (3, 'Charlie', 28, 'Marketing')
]
cursor.executemany('INSERT INTO ??? VALUES (?, ?, ?, ?)', employees) # TODO: 补充要插入的表名
# 查询所有记录
cursor.execute('SELECT * FROM employees')
rows = cursor.fetchall()
# 打印输出
for row in rows:
    print(row)
# 提交事务并关闭连接
conn.commit()
cursor.close()
conn.close()
```
